## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Ikan Betta ===
data = {
    "Panjang Tubuh (cm)": [6.0, 5.8, 6.2, 5.5, 6.3, 5.7, 6.1,
                           5.0, 5.1, 4.8, 5.2, 5.0, 4.9, 5.3, 5.1],
    "Berat Tubuh (g)": [3.0, 2.9, 3.2, 2.5, 3.3, 2.8, 3.1,
                        2.0, 2.1, 1.8, 2.2, 2.0, 1.9, 2.3, 2.1],
    "Kelas": [
        "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens",
        "Betta splendens", "Betta splendens",
        "Betta imbellis", "Betta imbellis", "Betta imbellis", "Betta imbellis", "Betta imbellis",
        "Betta imbellis", "Betta imbellis", "Betta imbellis"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding (imbellis=0, splendens=1) ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])

# === Fitur dan Target ===
X = df[["Panjang Tubuh (cm)", "Berat Tubuh (g)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## Visualisasi Splitting

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data: Panjang & Berat 15 Ikan Betta ===
data = {
    "Panjang Tubuh (cm)": [5.5, 5.8, 6.2, 5.5, 6.3, 5.7, 6.1, 5.0, 5.1, 4.8, 5.2, 5.0, 4.9, 5.3, 5.1],
    "Berat Tubuh (g)":     [2.2, 2.4, 2.8, 2.3, 3.0, 2.5, 2.9, 1.9, 2.0, 1.7, 2.1, 1.8, 1.8, 2.2, 2.0],
    "Kelas": [
        "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens",
        "Betta splendens", "Betta splendens",
        "Betta imbellis", "Betta imbellis", "Betta imbellis", "Betta imbellis",
        "Betta imbellis", "Betta imbellis", "Betta imbellis", "Betta imbellis"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # splendens = 1, imbellis = 0

# === Fitur dan Target ===
X = df[["Panjang Tubuh (cm)", "Berat Tubuh (g)"]]
y = df["Label"]

# === Latih Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === Cetak Pohon Keputusan ===
print("\n=== Pohon Keputusan ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Panjang Tubuh (cm)': [5.0, 6.0],
    'Berat Tubuh (g)': [1.9, 2.7]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (p, b) in enumerate(zip(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (g)'])):
    print(f"Data {i+1}: Panjang={p}, Berat={b}g â†’ Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 0.01

x_min, x_max = X['Panjang Tubuh (cm)'].min() - 0.2, X['Panjang Tubuh (cm)'].max() + 0.2
y_min, y_max = X['Berat Tubuh (g)'].min() - 0.2, X['Berat Tubuh (g)'].max() + 0.2

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Tubuh (cm)', 'Berat Tubuh (g)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Tubuh (cm)'], X.iloc[idx]['Berat Tubuh (g)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.xlabel("Panjang Tubuh (cm)")
plt.ylabel("Berat Tubuh (g)")
plt.title("Decision Boundary (Betta splendens vs Betta imbellis)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

# === 1. Dataset Cupang ===
data = {
    "Panjang Tubuh (cm)": [5.2, 5.5, 5.0, 5.1, 5.3, 4.9, 5.4, 5.2, 5.3, 5.0, 4.2, 4.3, 4.5, 4.4, 4.1],
    "Berat Tubuh (g)":    [2.0, 2.2, 1.9, 2.1, 2.3, 1.8, 2.4, 2.1, 2.3, 1.9, 1.5, 1.6, 1.7, 1.4, 1.5],
    "Kelas": [
        "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens",
        "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens", "Betta splendens",
        "Betta imbellis", "Betta imbellis", "Betta imbellis", "Betta imbellis", "Betta imbellis"
    ]
}
df = pd.DataFrame(data)

# === 2. Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # splendens=1, imbellis=0

# === 3. Fitur dan Target ===
X = df[["Panjang Tubuh (cm)", "Berat Tubuh (g)"]]
y = df["Label"]

# === 4. Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 5. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === 6. Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Panjang Tubuh (cm)': [4.6, 5.1],
    'Berat Tubuh (g)': [1.6, 2.2]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (p, b) in enumerate(zip(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (g)'])):
    print(f"Data {i+1}: Panjang={p} cm, Berat={b} g â†’ Prediksi: {hasil[i]}")

# === 7. Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 0.01

x_min, x_max = X['Panjang Tubuh (cm)'].min() - 0.2, X['Panjang Tubuh (cm)'].max() + 0.2
y_min, y_max = X['Berat Tubuh (g)'].min() - 0.2, X['Berat Tubuh (g)'].max() + 0.2

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Panjang Tubuh (cm)', 'Berat Tubuh (g)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Panjang Tubuh (cm)'], X.iloc[idx]['Berat Tubuh (g)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji
plt.scatter(uji_baru['Panjang Tubuh (cm)'], uji_baru['Berat Tubuh (g)'],
            c='black', marker='x', s=100, label='Data Uji')

# Tambahkan label untuk titik uji
for i in range(len(uji_baru)):
    plt.text(uji_baru['Panjang Tubuh (cm)'][i] + 0.02,
             uji_baru['Berat Tubuh (g)'][i] + 0.02,
             f"Uji {i+1}", fontsize=9, color='black')

plt.xlabel("Panjang Tubuh (cm)")
plt.ylabel("Berat Tubuh (g)")
plt.title("Decision Boundary: Betta splendens vs Betta imbellis")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
